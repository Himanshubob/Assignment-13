# Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example

# A data manipulation language (DML) is a family of computer languages including commands permitting users to manipulate data in a database. This manipulation involves inserting data into database tables, retrieving existing data, deleting data from existing tables and modifying existing data. DML is mostly incorporated in SQL databases

# INSERT=Used to insert new data records or rows in the database table
# UPDATE=Used to set the value of a field or column for a particular record to a new value
# DELETE=Used to remove one or more rows from the database table

# Q5. Explain Primary Key and Foreign Key.

# ans..A primary key is used to ensure that data in the specific column is unique. A column cannot have NULL values. It is either an existing table column or a column that is specifically generated by the database

# A foreign key is a column or group of columns in a relational database table that provides a link between data in two tables. It is a column (or columns) that references a column (most often the primary key) of another table.
